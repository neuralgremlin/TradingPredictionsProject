```python
pd.set_option('display.max_colwidth', None)
data.head()
```

```python

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
emptyline =[]

for row in data['title']:
  vs = analyzer.polarity_scores(row)
  emptyline.append(vs)
```

```python
emptyline[0:50]
```

```python
#Convertir el resultado de vader sentiment a un dataframe, para añadirlo a mis datos
sentiments = pd.DataFrame(emptyline)
#Agregando las columnas de vader
data['neg'] = sentiments['neg']
data['neu'] = sentiments['neu']
data['pos'] = sentiments['pos']
data['compound'] = sentiments['compound']
data.sort_values('date')
```

## como hay días que tienen mas de una noticia, quisieramos el sentimiento general de las noticias de ese día (para graficar 1 solo punto) vamos a convetir esta tabla:

```python
import numpy as np
df = pd.pivot_table( data, values=['neg','neu','pos','compound'], index=['Stock', 'date'],
                    aggfunc= np.mean)
```

```python
df.head()
```

```python
df.reset_index(level=0, inplace=True)
df.reset_index(level=0, inplace=True)# va 2 veces porque tiene 2 indexes de la tabla pivote
df.Stock.ffill(inplace=True)
```

```python
df.head()
```

#Seguramente la columna que queremos graficar es la de "Compound" que es la suma de positivo mas negativo (y lo neutral queda en ceros)
